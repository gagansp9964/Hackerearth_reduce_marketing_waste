```python
df_test = pd.read_csv('test.csv')
df_test.head()
```

```python
df_test.info()
```

```python
df_test.isnull().sum()
```

```python
df_test['Deal_value'] = df_test['Deal_value'].str.replace('$', '')
df_test['Weighted_amount'] = df_test['Weighted_amount'].str.replace('$', '')
```

```python
df_test['Deal_value']=df_test['Deal_value'].astype('float')
df_test['Weighted_amount']=df_test['Weighted_amount'].astype('float')
df_test.head()
```

```python
df_test.drop(["Lead_name","Industry","Date_of_creation","Contact_no","Location","POC_name","Designation","Lead_POC_email","Hiring_candidate_role","Lead_source","Last_lead_update","Internal_POC","Resource"], axis=1, inplace=True)
```

```python
p=pd.get_dummies(df_test['Pitch'],drop_first=True)
lr=pd.get_dummies(df_test['Lead_revenue'],drop_first=True)
fc=pd.get_dummies(df_test['Fund_category'],drop_first=True)
g=pd.get_dummies(df_test['Geography'],drop_first=True)
lm=pd.get_dummies(df_test['Level_of_meeting'],drop_first=True)
df_test=pd.concat([df_test,p,lr,fc,g,lm],axis=1)
df_test.drop(["Pitch","Lead_revenue","Fund_category","Geography","Level_of_meeting"],axis=1,inplace=True)
```

```python
df_test.head()
```

```python
df=pd.DataFrame(df_test,columns=["Deal_title"])
```

```python
sb.heatmap(df_test.isnull(),yticklabels=False,cmap='viridis')
```

```python
df_test['Deal_value'] = df_test['Deal_value'].fillna((df1['Deal_value'].mean()))
df_test['Weighted_amount'] = df_test['Weighted_amount'].fillna((df1['Weighted_amount'].mean()))
df_test['Internal_rating'] = df_test['Internal_rating'].fillna((df1['Internal_rating'].mean()))
df_test.head() 
```

```python
df_test.drop(["Deal_title"],axis=1,inplace=True)
```

```python
sb.heatmap(df_test.isnull(), yticklabels=False, vmin=0, vmax=1)
```

```python
x_train = df1.drop(['Success_probability'], axis=1)
y_train = df1['Success_probability']
x_test=df_test
```

```python
import xgboost as xgb
xg_reg = xgb.XGBRegressor(objective ='reg:squarederror', colsample_bytree = 0.3, learning_rate = 0.1,
                max_depth = 5, alpha = 10, n_estimators = 10)
```

```python
xg_reg.fit(x_train,y_train)
xg_pred = xg_reg.predict(x_test)
df["Success_probability"]=xg_pred
df.head()
```
